```
import pandas as pd
import os
import json
import csv
import plotly.express as px
import numpy as np
```

**To get a aggregated transaction details in csv format**

```
path = "/content/pulse/data/aggregated/transaction/country/india/state"
agg_trans = os.listdir(path)

agg_data = {'State':[], 'Year':[], 'Quarter':[], 'Transaction_type':[], 'Transaction_count':[], 'Transaction_amt':[]}

for i in agg_trans:
  state_i = os.path.join(path,i)
  for j in os.listdir(state_i):
    year_j = os.path.join(state_i,j)
    for k in os.listdir(year_j):
      quarter_k = os.path.join(year_j,k)
      with open(quarter_k, 'r') as data_file:
        D = json.load(data_file)
        for item in D['data']['transactionData']:
          Name = item['name']
          Count = item['paymentInstruments'][0]['count']
          Amount = item['paymentInstruments'][0]['amount']
          agg_data['State'].append(i)
          agg_data['Year'].append(j)
          agg_data['Quarter'].append(int(k.strip('.json')))
          agg_data['Transaction_type'].append(Name)
          agg_data['Transaction_count'].append(Count)
          agg_data['Transaction_amt'].append(Amount)

agg_trans_dataframe = pd.DataFrame(agg_data)
```

**To get aggregated user details in csv format**

```
path = "/content/pulse/data/aggregated/user/country/india/state"
agg_user = os.listdir(path)

agg_data = {'State':[], 'Year':[], 'Quarter':[],'Device_brand':[],'Count':[],'Percentage':[]}

for i in agg_user:
  state_i = os.path.join(path,i)
  for j in os.listdir(state_i):
    year_j = os.path.join(state_i,j)
    for k in os.listdir(year_j):
      quarter_k = os.path.join(year_j,k)
      with open(quarter_k, 'r') as data_file:
        D = json.load(data_file)
        try:
          for item in D['data']['usersByDevice']:
            Brand = item['brand']
            Count = item['count']
            Percentage = item['percentage']
            agg_data['State'].append(i)
            agg_data['Year'].append(j)
            agg_data['Quarter'].append(int(k.strip('.json')))
            agg_data['Device_brand'].append(Brand)
            agg_data['Count'].append(Count)
            agg_data['Percentage'].append(Percentage)
        except:
          pass
agg_user_dataframe = pd.DataFrame(agg_data)
```
